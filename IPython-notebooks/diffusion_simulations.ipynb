This is a notebook with the numerical diffusion method implemented. In this very simplified model, each 'neuron' in this case is randomly populated in a 2D grid, and is solely characterised by its constant rate of NO synthesis. 

Different boundary conditions and diffusion paremeters may be explored. 

```python
import numpy
from numpy import arange as arange
import pylab

dt_diff = 1 # *ms (use ms units if using brian)


NO_diff = 1.0e-6   # diffusion constant for NO # mm^2/ms = 3300 micron^2/s
NOdecay_rate = 1.0e-4 # decay rate for NO ( ms^-1)

ds = 0.005
size = 1.0
time = 15000

t_off = 20000

dx2 = ds**2
dy2 = ds**2
dt_diff = dx2*dy2/( 2*NO_diff*(dx2+dy2) )

#zero boundary and initial conditions
NO_concs = numpy.zeros((int(size/ds),int(size/ds)))

#NO_concs = numpy.ones((int(size/ds),int(size/ds)))*0.15

## populates grid with neurons at a probability of 0.02
neuron_present = numpy.random.binomial(2,0.02,(int(size/ds),int(size/ds)))

## assigns random NO production rate to each neuron; 0<rate<1
NO_rates =  numpy.random.random((int(size/ds),int(size/ds)))*neuron_present

nx = int(size/ds)
ny = int(size/ds)

neumann_grad = 1e-6
```

```python
snapshot_resolution = 1000

NO_mean_dirichlet = numpy.zeros(time)
NO_std_dirichlet = numpy.zeros(time)

NO_snapshots_dirichlet = numpy.zeros((int(time/snapshot_resolution),int(size/ds),int(size/ds)))

NO_concs = numpy.zeros((int(size/ds),int(size/ds)))

for t in xrange(time):
    ## direchlet boundary conditions
    NO_concs[1:-1, 1:-1] = NO_concs[1:-1, 1:-1] + dt_diff*(NO_diff*( (NO_concs[2:, 1:-1] - 2*NO_concs[1:-1, 1:-1] + NO_concs[:-2, 1:-1])/dx2 + (NO_concs[1:-1, 2:] - 2*NO_concs[1:-1, 1:-1] + NO_concs[1:-1, :-2])/dy2 ) + NO_rates[1:-1, 1:-1] - NOdecay_rate*NO_concs[1:-1, 1:-1])

    NO_mean_dirichlet[t] = numpy.mean(NO_concs)
    NO_std_dirichlet[t] = numpy.std(NO_concs)

    if t%snapshot_resolution == 0:
        print 't = ', t
        NO_snapshots_dirichlet[int(t/snapshot_resolution)] = NO_concs.copy()

    
NO_mean_pbc = numpy.zeros(time)
NO_std_pbc = numpy.zeros(time)

NO_snapshots_pbc = numpy.zeros((int(time/snapshot_resolution),int(size/ds),int(size/ds)))

NO_concs = numpy.zeros((int(size/ds),int(size/ds)))
for t in xrange(time):
    ## periodic boundary conditions
    leftX = arange(nx)-1
    rightX = (arange(nx)+1)%nx
    leftY = arange(ny)-1
    rightY = (arange(ny)+1)%ny
    NO_concs = NO_concs + dt_diff*(NO_diff*((NO_concs[rightX,:]-2*NO_concs + NO_concs[leftX,:])/dx2 + (NO_concs[:,rightY] - 2*NO_concs
        + NO_concs[:,leftY])/dy2)+ NO_rates - NOdecay_rate*NO_concs)

    NO_mean_pbc[t] = numpy.mean(NO_concs)
    NO_std_pbc[t] = numpy.std(NO_concs)
    
    if t%snapshot_resolution == 0:
        print 't = ', t
        NO_snapshots_pbc[int(t/snapshot_resolution)] = NO_concs.copy()    

        
#  neumann bc from http://ipython-books.github.io/featured-05/        

NO_mean_neumann = numpy.zeros(time)
NO_std_neumann = numpy.zeros(time)

NO_snapshots_neumann = numpy.zeros((int(time/snapshot_resolution),int(size/ds),int(size/ds)))

NO_concs = numpy.zeros((int(size/ds),int(size/ds)))

NO_rates =  numpy.random.random((int(size/ds),int(size/ds)))*neuron_present

for t in xrange(time):
    ## neumann boundary conditions

    
    NO_concs[1:-1, 1:-1] = NO_concs[1:-1, 1:-1] + dt_diff*(NO_diff*( (NO_concs[2:, 1:-1] - 2*NO_concs[1:-1, 1:-1] + NO_concs[:-2, 1:-1])/dx2 + (NO_concs[1:-1, 2:] - 2*NO_concs[1:-1, 1:-1] + NO_concs[1:-1, :-2])/dy2 ) + NO_rates[1:-1, 1:-1] - NOdecay_rate*NO_concs[1:-1, 1:-1])

    NO_concs[0,:] = NO_concs[1,:]
    NO_concs[-1,:] = NO_concs[-2,:]
    NO_concs[:,0] = NO_concs[:,1]
    NO_concs[:,-1] = NO_concs[:,-2]
    
    NO_mean_neumann[t] = numpy.mean(NO_concs)
    NO_std_neumann[t] = numpy.std(NO_concs)
    
    if t%snapshot_resolution == 0:
        print 't = ', t
        NO_snapshots_neumann[int(t/snapshot_resolution)] = NO_concs.copy()
        
```

#Different boundary conditions

dirichlet: zero concentration at boundary

periodic: .. periodic

nuemann: zero gradient at boundary
    
Periodic and neumann are very similar, but neumann is more stable (avoid instability issues in simulations anyway)

If you take inner square of dirichlet simulations, this is also very similar to neumann and periodic boundaries


```python
%matplotlib inline
pylab.fill_between(range(len(NO_mean_dirichlet)),pylab.add(NO_mean_dirichlet,NO_std_dirichlet),pylab.subtract(NO_mean_dirichlet,NO_std_dirichlet), facecolor='blue', alpha=0.5)
pylab.plot(NO_mean_dirichlet,'b',label='dirichlet')
pylab.fill_between(range(len(NO_mean_pbc)),pylab.add(NO_mean_pbc,NO_std_pbc),pylab.subtract(NO_mean_pbc,NO_std_pbc), facecolor='green', alpha=0.5,label='periodic bc')
pylab.plot(NO_mean_pbc,'g',label='periodic bc')
pylab.fill_between(range(len(NO_mean_neumann)),pylab.add(NO_mean_neumann,NO_std_neumann),pylab.subtract(NO_mean_neumann,NO_std_neumann), facecolor='red', alpha=0.5,label='neumann')
pylab.plot(NO_mean_neumann,'r',label='neumann')
pylab.legend()
pylab.ylabel('NO concentration')
pylab.xlabel('time (ms)')
pylab.show()
```
